## **Data Visualization**

- **Pandas:** Built on matpltlib, simplifies quick ploting directly from DataFrames.
- **Seaborn:** Built on Matplotlib, simplifies complex statistical visualizations and improves the aesthetics of the plot.
- **Matplotlib:** The foundation of both Pandas and Seaborn visualizations. It provides the core plotting functions, but with more flexibility at the cost of increased complexity.

*When to Use Each Library:*

- **Pandas:** When you need quick basic visualizations directly from your DataFrame.
- **Matplotlib:** When you need fine control and customization of plots or when creating more complex visualizations.
- **Seaborn:** When you want aesthetically pleasing plots with less code and when you're focusing on statistical relationships in the data.

### *Most commonly used visualization*
- **Bar Chart:** Categorical comparison using rectangular bars.
  - **When to Use:** Comparing different groups or categories (e.g sales by region, product br popularity).
  - **Use For:** Dicrete categories; when you have a few categories.
    
- **Histogram:** Description of a numerical variables using bins.
  - **When to Use:** To observe the frequency distribution of data (e.g age distribution, income ranges).
  - **Use For:** Understanding the spread or skewness of data.
    
- **Line Chart:** usually for time-series data.
  - **When to Use:** To show trends or changes over time (e.g stock prices over month).
  -  **Use For:** Tracking changes over a continuous interval (like time).
    
- **Scatter Plot:** Points plotted on two axes to display the relationship between two variables.
  - **Where to Use:** To show correlation or relationships between two numeric variables (Sales vs Profit).
  - **Use For:** Identifying relationships, clusters or outliers.
    
- **Pie Chart:** Circular charts divided into slices to represent proportions.
  - **When to Use:** To show parts of a whole (e.g market share, budget allocation).
  - **Use For:** Simple proportion comparison; avoid with many categories.
    
- **Boxplot:** Displays distribution, quartiles, and outliers of data.
  - **When to Use:** To compare distributions across categories (e.g profit across different product lines).
  - **Use For:** Visualizing spread, outliers, and comparing distributions, particularly with statistical data.

## **Outlier**

- Outliers are data that deviate significantly from the rest of the dataset. They can be much higher or lower than most of the values.
- Example: In a dataset of salaries, if most salaries are between $30,000 and $60,000, but one salary is $500,000, this is an outlier.

***What are outliers important?***
- **Influence on Results:** Outliers can distort statistical measures (like the mean) and affect machine learning models by leading to biased or inaccurate results.
- **Data Quality:** Outliers might indicate errors in data collection, incorrect data entry, or rare but important cases.

***How to Detect Outliers***
- **Boxplots:** Outliers are often represented as points outside the whiskers of a boxplot.
- **Histogram:** Extremely low or high values will appear isolated from the rest of the data.
- **ScatterPlots:** For bivariate data, points that are far from the general clusters can be outliers.
- **IQR (Interquartile Range):** Calculate IQR as the difference between the 75th percentile (Q3) and the 25th percentile (Q1). Outliers are typically data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.



***Handling Outliers:**
- **Keep them:** Sometimes, outliers are valid data points that represent rare but real phenomena. For example, a company with unusually high sales could represent a valid and valuable case.
- **Remove them:** If outlier results from errors or noise, removing them can lead to more accurate analysis.
- **Cap Them:** In some cases, you can "cap" outliers by replacing extreme values with maximum or minimum thresholds (known as winsorizing).
- **Transform Data:** Use log transformations or other techniques to reduce the influence of outliers while keeping them in the dataset. 

## **Detecting and Handling Outliers in Python**
- Boxplot example

```python
sd.boxplot(column = "Profit")
plt.title("Boxplot for Profit")
plt.show()
```

```python
sns.scatterplot(x = "Sales", y = "Profit", data = sd, color = cm.coolwarm(1))
```

```python
# Calculate Q1 and Q3
Q1 = sd["Profit"].quantile(0.25)
Q3 = sd["Profit"].quantile(0.75)
IQR = Q3 - Q1

print(Q1, "\n")
print(Q3, "\n")
print(IQR)
```

```python
# Filter Outliers
# outliers = sd[(sd["Profit"] < (Q1 - 1.5 * IQR)) | (sd["Profit"] > (Q1 + 1.5 * IQR))]
```

```python
# Checking for outliers using IQR
# by creating a lower bound and upper bound and checking for values that falls belows the lower bound and above the upper bound.
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q1 + 1.5 * IQR

print(lower_bound, "\n")
print(upper_bound)
outliers = sd[(sd["Profit"] < lower_bound) | (sd["Profit"] > upper_bound)]
```

**Why and When to Handle Outliers?**
- **When Outliers Distort Models:** Machine learning models, especially those based on regression or distance metrics like (KNN), can be highly sensitive to outliers.
- **When Outliers Are Errors:** If outliers are due to data entry mistakes or sensor malfunction, removing or correcting them is important for data integrity.
