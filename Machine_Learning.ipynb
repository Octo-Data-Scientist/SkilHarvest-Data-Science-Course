## Machine Learning:
- A subset of artificial intelligence (AI) that enables a system to learn from data and improve it's performance over time without being explicitly programmed.

### Three main types of ML:
- **Supervised Learning:** The model is trained on **labeled data**. It means both the input and output are known.

    - *Two types* - **Regression**(target values are continuous numeric variables - e.g. predicting house prices) and **classification** (target value is a categorical variable e.g. predicting the house or discrete).


- **Unsupervised Learning:** The model learns from unlabeled data, finding hidden patterns or groupings. Examples: Clustering customers, dimensionality reduction.

  
- **Reinforcement Learning:** The model learns by interacting with an environment and getting feedback in the form of rewards or penalties.

### Key Terms in Machine Learning:
- **Training Data:** The dataset used to teach the model. The model learns patterns from this data to make predictions.
  
- **Testing Data:** A separate dataset used to evaluate the model's performance after training. It shows how well the model works on new unseen data.

  
- **Features:** These are the input variables or columns in the dataset used to make predictions. For example, in a dataset predicting house prices, features might include "size of the house," "number of rooms," and "location."

- **Labels (or Target):** This is the output variable the model is trying to predict. In a house prices dataset, the label would be the actual "price" of the house.

- **Model:** A model is built using data and algorithms. It is a program trained to recognize patterns or relationships in data and make predictions or decisions based on patterns on new, unseen data.

- **Algorithm:** defines the rules the model uses to find patterns in the data.

- **Overfitting:** happens when a model learns too much from the training data, capturing noise and specific details that don't generalize well to new data. Essentially, the model becomes too complex, "memorizing" the training set rather than understanding the general patterns.

    - **Signs of overfitting**
        - a. High accuracy on training data but poor accuracy on testing data.
        - b. The model performs very well on the data it was trained on but fails to predict well on new data.
    - **Fixing Overfitting**: Some ways to fix overfitting -
      - a. Simplify the Model- Use a simpler model with fewer features or less complex algorithms.
      - b. Cross-Validation: test the model on different subsets of the data.
      - c. Get More Data: add more training data. With more examples, the model has a better chance of learning general patterns                               rather than specifics.
      - d. Regularization: Add a penalty to the model's complexity, which discourages the model from relying heavily on specific                                features .
- **Underfitting:** occurs when a model is too simple to capture the underlying patterns in the data. It may struggle to learn from the training data and perform poorly on the training and testing data.
    - **Signs**
      - Low accuracy on both the training and testing sets.
      - The model fails to capture trends even on the training data, suggesting it's too simplistic.
    - **Fixing Underfitting:**
      - Increase Model Complexity: Use a more complex model.
      - Add More Features: If possible, introduce additional features or transformations that might help capture more information                               from the data.
      - Reduce Regularization: If you're using regularization, reduce it slightly so the model has more flexibility to fit the data.

## Key Steps in Machine Learning
- **Data Collection:** Gather or load data you'll use for training.
- **Data Preprocessing:** Clean and transform the data to make it suitable for analysis.
- **Model Selection:** Choose an algorithm to fit the data.
- **Training:** Feed the model with data and teach it to make predictions.
- **Evaluation:** Test the model to check how well it performs.
- **Tuning:** Improve model performance by adjusting parameters.
